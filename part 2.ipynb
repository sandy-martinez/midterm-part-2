```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
countries = pd.read_csv('countries.csv')
```

This data set has recorded population of countries in 5-year increments from 1952 to 2007.  

```python
countries
```

I will compare population growth between the US, a **developed** country; and India, a **developing** country that is experiencing high economic growth.

```python
us = countries[countries.country == 'United States']
```

```python
us
```

```python
india = countries[countries.country == 'India']
```

```python
india
```

Now I will plot US population in millions:

```python
plt.plot(us.year, us.population / 10**6)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('US Population')
plt.show()
```

Now I will plot India population in millions:

```python
plt.plot(india.year, india.population / 10**6)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('India Population')
plt.show()
```

US and India population:

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(india.year, india.population / 10**6)
plt.legend(['United States', 'India'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('US and India Population')
plt.show()
```

Based on the graph, India seems to be experiencing faster population growth compared to the US's relatively stable growth.

Now I will compare the **percentage** growth between the two countries:

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(india.year, india.population / india.population.iloc[0] * 100)
plt.legend(['United States', 'India'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('US and India Population Growth')
plt.show()
```

Based on the graph, India's population is growing at a faster rate compared to the US, especially in the later years.
